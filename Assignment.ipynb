3) Explain the following terms:

  . The difference in np.array, np.asarray and np.asanyarray.
  
  . The difference between Deep copy and shallow copy.

Ans) 
    np.array: 

    It creates a new array from any object exposing the array interface, or from any nested sequence.

    np.asarray:
    
    Converts the input to an array, but does not copy data if the input is already an ndarray (or subclass of ndarray).

    np.asanyarray:
        
     Converts the input to an array, but passes through subclasses of ndarray (like matrix).

difference between shallow copy and deep copy

shallow copy:
    
    A shallow copy creates a new object, but inserts references into it to the objects found in the original.

Deep copy:
    
     A deep copy creates a new object and recursively copies all objects found in the original.

17) Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import seaborn as sns
```

```python
x = np.random.randn(50)
y = np.random.randn(50)

quadrants = []
for xi, yi in zip(x, y):
    if xi >= 0 and yi >= 0:
        quadrants.append('Q1')
    elif xi < 0 and yi >= 0:
        quadrants.append('Q2')
    elif xi < 0 and yi < 0:
        quadrants.append('Q3')
    else:
        quadrants.append('Q4')
        
data = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='bright')




plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.title('Quadrant-wise Scatter Plot')

plt.legend(title='Quadrant')
plt.show()
```

19) Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.

```python
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Output to notebook
output_notebook()

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a new plot
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values')

# Add bars with color based on their values
colors = ['#%02x%02x%02x' % (int(r), int(g), 150) for r, g in zip(50+2*values, 30+2*values)]
p.vbar(x='categories', top='values', width=0.9, color=colors, legend_field='categories', source=source)

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Set title and labels
p.title.text = 'Random Categorical Bar Chart'
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Show the plot
show(p)
```
